```python
# Define valid ranges
valid_ranges = {
    'BMI': (15, 40),
    'Measles': (0, 1000),
    'infant deaths': (0, 1000),
    'under-five deaths': (0, 1000)
}

# Loop through each column
for col in valid_ranges:
    
    # Get min and max values for that column
    min_val = valid_ranges[col][0]
    max_val = valid_ranges[col][1]
    
    # Loop through each value in the column
    new_values = []
    
    for value in df[col]:
        
        if value < min_val:
            new_values.append(min_val)
        
        elif value > max_val:
            new_values.append(max_val)
        
        else:
            new_values.append(value)
    
    # Replace column with cleaned values
    df[col] = new_values

# Show summary
print(df[['BMI', 'Measles', 'infant deaths', 'under-five deaths']].describe())
```

# Encoding

```python
le=LabelEncoder()
df['Status']=le.fit_transform(df['Status'])
df.head()
```

# ML

```python
# Remove unwanted columns
df = df.drop(['Year', 'Measles', 'Population', 'Country'], axis=1)
```

```python
df.head()
```

```python
# feature selection
x = df.drop('Life expectancy',axis=1)
y = df['Life expectancy']
```

# LinearRegression

```python
# train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(y_test.shape)

# Model training
lr=LinearRegression()
lr.fit(x_train,y_train)

# prediction
y_pred=lr.predict(x_test)
y_pred

# evaluate
r2 = r2_score(y_test,y_pred)
print("LinearRegression:",r2)
```

# RandomForestRegressor

```python
# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_test.shape)

# Model training
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)

# prediction
y_pred = rf.predict(x_test)

# evaluate
r2 = r2_score(y_test, y_pred)
print("Random Forest R2:", r2)
```

# DecisionTreeRegressor

```python
# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_test.shape)

# Model training
dt = DecisionTreeRegressor(random_state=42)
dt.fit(x_train, y_train)

# prediction
y_pred = dt.predict(x_test)


# evaluate
r2 = r2_score(y_test, y_pred)
print("Decision Tree R2:", r2)
```

# SVR

```python
# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_test.shape)

# Scaling 
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Model training
svr = SVR(kernel='rbf')
svr.fit(x_train_scaled, y_train)

# prediction
y_pred = svr.predict(x_test_scaled)

# evaluate
r2 = r2_score(y_test, y_pred)
print("Scaled SVR R2:", r2)

```

# KNN Regressor

```python
# train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_test.shape)

# Scaling 
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Model training
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train_scaled, y_train)

# prediction
y_pred = knn.predict(x_test_scaled)

# evaluate
r2 = r2_score(y_test, y_pred)
print("KNN Regressor R2:", r2)
```
